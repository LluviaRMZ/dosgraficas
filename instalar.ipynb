```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.show

plt.plot(x ,y, label= 'x^2')
plt.title('Simple Line Plot')
plt.xlable('x Axis')
plt.ylable('y Axis')
plt.show()
```

```python
categoria = ['A', 'B','C']
Values = [7,3,5]
plt.bar(categoria, Values)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show
```
